##1) _updated_by

```python
#Obtengo un análisis univariado de una variable categórica en específico
table1 = freq_tbl(Filtro20['_updated_by'])
table1
```

```python
#Obtengo un fultro de los valores más relevantes de la variable categórica seleccionada
Filtro1_1 = table1[table1['frequency']>5]
Filtro1_1
```

```python
#Ajusto el índice de mi DataFrame
Filtro1_index = Filtro1_1.set_index('_updated_by')
Filtro1_index
```

```python
#Realizamos gráfico de barras del dataframe filtrado
Filtro1_index.plot(kind = 'bar',width= 1,figsize=(10,5))
plt.title('Desarrollo de entrevistas')
plt.xlabel('Entrevistador')
plt.ylabel('Frecuencia')
```
